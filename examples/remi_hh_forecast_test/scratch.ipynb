```python
import pandas as pd
```

```python
pums_hh = pd.read_csv('data/seed_households.csv')
```

```python
synth_hh = pd.read_csv('output/synthetic_households.csv')
```

```python
pums_hh['income_bin'] = pd.cut(
    pums_hh['HINCP'],
    bins=[-1, 49999, 99999, 149999, float('inf')],
    labels=['Under 50k', '50k-100k', '100k-150k', '150k+']
)
```

```python
pums_income = pums_hh.groupby('income_bin')['WGTP'].sum()
```

```python
synth_hh['income_bin'] = pd.cut(
    synth_hh['HINCP'],
    bins=[-1, 49999, 99999, 149999, float('inf')],
    labels=['Under 50k', '50k-100k', '100k-150k', '150k+']
)

synth_income = synth_hh.groupby('income_bin').size()
```

```python
# Display results in a bar plot
import matplotlib.pyplot as plt
income_df = pd.DataFrame({
    'PUMS': pums_income,
    'Synthetic': synth_income
})

income_df.plot(kind='bar')
plt.ylabel('Household Count')
plt.title('Income Distribution Comparison')
plt.show()
```

```python
pums_hh['hhsz_bin'] = pd.cut(
    pums_hh['NP'],
    bins=[0, 1, 2, 3, 4, float('inf')],
    labels=['hhsz_1', 'hhsz_2', 'hhsz_3', 'hhsz_4', 'hhsz_5+']
)

pums_hhsz = pums_hh.groupby('hhsz_bin')['WGTP'].sum()
```

```python
synth_hh['hhsz_bin'] = pd.cut(
    synth_hh['NP'],
    bins=[0, 1, 2, 3, 4, float('inf')],
    labels=['hhsz_1', 'hhsz_2', 'hhsz_3', 'hhsz_4', 'hhsz_5+']
)

synth_hhsz = synth_hh.groupby('hhsz_bin').size()
```

```python
hhsz_df = pd.DataFrame({
    'PUMS': pums_hhsz,
    'Synthetic': synth_hhsz
})

hhsz_df.plot(kind='bar')
plt.ylabel('Household Count')
plt.title('Household Size Distribution Comparison')
plt.show()
```

```python
pums_hh['vehicle_bin'] = pd.cut(
    pums_hh['VEH'],
    bins=[-1, 0, 1, 2, float('inf')],
    labels=['No Vehicle', '1 Vehicle', '2 Vehicles', '3+ Vehicles']
)

pums_vehicle = pums_hh.groupby('vehicle_bin')['WGTP'].sum()
```

```python
synth_hh['vehicle_bin'] = pd.cut(
    synth_hh['VEH'],
    bins=[-1, 0, 1, 2, float('inf')],
    labels=['No Vehicle', '1 Vehicle', '2 Vehicles', '3+ Vehicles']
)

synth_vehicle = synth_hh.groupby('vehicle_bin').size()
```

```python
vehicle_df = pd.DataFrame({
    'PUMS': pums_vehicle,
    'Synthetic': synth_vehicle
})

vehicle_df.plot(kind='bar')
plt.ylabel('Household Count')
plt.title('Vehicle Ownership Distribution Comparison')
plt.show()
```
